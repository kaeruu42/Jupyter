```python
import math
import matplotlib.pyplot as plt
import numpy
from scipy import stats

sinDict = {}
```

```python
def sinMaker():
    x = 0
    while x < 100:
        sinDict.update({x: math.sin(x)})
        x += .01
    
```

```python
sinMaker()
print(sinDict)
```

```python
lists = sorted(sinDict.items())
x, y = zip(*lists)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show
```

```python
# Linear Regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.plot(x, y)
plt.plot(x, mymodel, color = 'r')
plt.show()
```

```python
print(myfunc(20))
print(math.sin(20))
# Doesn't work
```

```python
# Polynomial Regression
polymodel = numpy.poly1d(numpy.polyfit(x, y, 100))

myline = numpy.linspace(0, 100, 10000)

plt.plot(x, y)
plt.plot(myline, polymodel(myline), color = 'r')
plt.show
```
